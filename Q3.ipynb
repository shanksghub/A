```python
import random # importing random module for generating pseduo random  number generators
import math # importing math for list of mathematical functions
number_hill_climbs = 200 # number of hill climbs which is the number of times the iterations are run
steps = 1000 # steps which denotes how many iterations to run
f_list = [] # empty list to make a list of fbest values.

for j in range(number_hill_climbs): # iterating for the number of hill climbs
    # returning random integers between numbers specified.
    a = random.randint(5,10)
    b = random.randint(5,10)
    c = random.randint(10,13)
    d = random.randint(11,17)
    e = random.randint(11,19)

    # Return the next random floating point number between [0.0, 1.0) 
    # after being multiplied by a constant which is the minimum duration of work.
    u = 5*random.random()
    v = 7*random.random()
    x = 10*random.random()
    y = 3*random.random()
    z = 7*random.random()
    fbest = 3*a*x + 5*b*y + 10*c*z + 7*u*d + 11*v*e # The equation to optimise.

    for i in range(1,steps): # Iterating over the number of steps
        # Adding an integer which is random from a range.
        an = random.randint(5,10) + random.randrange(6) 
        bn = random.randint(7,11) + random.randrange(5)
        cn = random.randint(10,13)+ random.randrange(4)
        dn = random.randint(5,11) + random.randrange(9)
        en = random.randint(7,13) + random.randrange(7)
        
        # Adding a value from Gaussian Distribution.
        un = u + random.gauss(0,0.001)
        vn = v + random.gauss(0,0.001)     
        xn = x + random.gauss(0,0.001)
        yn = y + random.gauss(0,0.001)
        zn = z + random.gauss(0,0.001)
                      
        # Constraints of the equation.              
        if((xn>=10 and xn<=13) and (yn>=3 and yn<=11) and (zn>=7 and zn<=13) \
           and (un>=5 and un<=10) and (vn>=7 and vn<=11) \
           and (xn + yn+zn+un+vn<=20) and (an>=5 and an<=10) and (bn>=5 and bn<=10) and (cn>=10 and cn<=13) \
           and (dn>=11 and dn<=17) and (en>=11 and en<=19) and (an+bn+cn+dn+en<=40)) :

            fnew = 3*an*xn + 5*bn*yn + 10*cn*zn + 7*un*dn +11*vn*en # Equation after changing values
            if fnew > fbest: # Comparing value of new and old equations and if new values are better, they become starting points.
                x = xn
                y = yn
                z = zn
                u = un
                v = vn
                
                a = an
                b = bn
                c = cn
                d = dn
                e = en
                fbest = fnew
      
    f_list.append(fbest) # Appending the best values of the equation.
   # Printing the values of the optimised equation.
    print ('xbest' , x, 'ybest' , y, 'zbest', z,'ubest'
           , u, 'vbest' , v, 'abest', an,'bbest', bn,'cbest'
           , cn,'dbest', dn,'ebest', en, 'fbest', fbest)
   
max(f_list) # returning the maximum from the list of fbest.
```
